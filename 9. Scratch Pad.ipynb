# Amphipod A Head Depth to Length Relationship

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

ama_x = []
ama_y = []
with open('data/AMA_HeadDepth_Length.csv', 'rb') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader, None)
    for row in reader:
        ama_x.append(float(row[0]))
        ama_y.append(float(row[1]))


(m, b) = np.polyfit(ama_x, ama_y, 1)
print "y = " + str(m) + "x " + "+ " + str(b)
yp = np.polyval([m,b], ama_x)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.set_xlabel("Length (mm)")
ax.set_ylabel("Head Depth (mm)")
ax.set_title("Amphipod AMA Length Head Depth Relationship")

ax.plot(ama_x, yp, color='r')
ax.scatter(ama_x, ama_y)
plt.show()
```
