# Amphipod A Head Depth to Length Relationship

```python
import csv
import matplotlib.pyplot as plt

ama_x = []
ama_y = []
with open('data/AMA_HeadDepth_Length.csv', 'rb') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader, None)
    for row in reader:
        ama_x.append(float(row[0]))
        ama_y.append(float(row[1]))

plt.figure()        
plt.xlabel = "Length (mm)"
plt.ylabel = "Head Depth (mm)"
plt.title = "Amphipod AMA Length Head Depth Relationship"
plt.scatter(ama_x, ama_y)
plt.show()
```
