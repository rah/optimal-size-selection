```python
import pandas as pd

df = pd.read_csv('../data/Benthos-Habitat-Complexity-25-26-6-83.csv')

df.head()
```

```python
df.fillna(0, inplace=True)
df.head()
```

```python
grouped = df.groupby('Category')
grouped.head()
```

```python
gmean = grouped.mean()
gmean.head()
```

```python
gmean.hist(column=['Seagrass Wt (gms)', 'HA', 'AMA', 'AMB'], by='Category')
```
